```python
#  TASK 2: Univariate Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Sales Data
sales = pd.DataFrame({
    'Product': ['A','B','C','A','B'],
    'Price': [200, 350, 120, 205, 360],
    'Quantity': [2,1,3,2,3],
    'Category': ['Electronics','Furniture','Stationery','Electronics','Furniture']
})

# --- Categorical Data ---
# Bar Chart
sns.countplot(x='Product', data=sales)
plt.title('Bar Chart  Product Count')
plt.show()

# Pie Chart
sales['Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart  Category')
plt.ylabel('')
plt.show()

# --- Continuous Data ---
# Histogram
sns.histplot(sales['Price'], kde=True)
plt.title('Histogram  Price Distribution')
plt.show()

# Strip Plot
sns.stripplot(x='Category', y='Price', data=sales, jitter=True)
plt.title('Strip Plot  Price per Category')
plt.show()

# Swarm Plot
sns.swarmplot(x='Category', y='Quantity', data=sales)
plt.title('Swarm Plot  Quantity per Category')
plt.show()
```
