```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Dataset
df = pd.DataFrame({
    'Category': ['Electronics','Furniture','Stationery','Electronics','Furniture'],
    'Price': [200, 350, 120, 205, 360],
    'Quantity': [2,1,3,2,3]
})

# Continuous vs Continuous  Scatter with Fit Line
sns.regplot(x='Price', y='Quantity', data=df, color='red')
plt.title('Scatter Plot with Fit Line')
plt.show()

# Categorical vs Continuous  Box Plot
sns.boxplot(x='Category', y='Price', data=df)
plt.title('Box Plot  Price per Category')
plt.show()

# Violin Plot
sns.violinplot(x='Category', y='Quantity', data=df)
plt.title('Violin Plot  Quantity per Category')
plt.show()

# Bar Plot (Category vs Average Price)
avg_price = df.groupby('Category')['Price'].mean().reset_index()
sns.barplot(x='Category', y='Price', data=avg_price)
plt.title('Bar Chart  Average Price per Category')
plt.show()
```
