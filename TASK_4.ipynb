```python
# 🧠 TASK 4: Multivariate Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Sample dataset
data = {'Date': pd.date_range(start='2024-01-01', periods=5),
        'Product': ['A','B','C','A','B'],
        'Price': [200,350,120,205,360],
        'Quantity': [2,1,3,2,3],
        'Region': ['North','South','East','West','North'],
        'Category': ['Electronics','Furniture','Stationery','Electronics','Furniture']}
df = pd.DataFrame(data)

# Pair Plot (Continuous Variables)
sns.pairplot(df[['Price','Quantity']])
plt.suptitle('Pair Plot – Continuous Variables', y=1.02)
plt.show()

# Line Plot (Quantity over Time)
sns.lineplot(x='Date', y='Quantity', hue='Product', data=df, marker='o')
plt.title('Line Plot – Quantity Over Time by Product')
plt.show()

# Stacked Bar (Region vs Category)
stacked = df.groupby(['Region','Category']).size().unstack()
stacked.plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart – Region vs Category')
plt.ylabel('Count')
plt.show()

# Plotly – Parallel Coordinates Plot
px.parallel_coordinates(df, color='Price',
                        dimensions=['Price','Quantity'],
                        title='Parallel Coordinates – Price & Quantity').show()
```
